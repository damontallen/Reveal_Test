```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D


points = np.ones(5)  # Draw 3 points for each line
text_style = dict(horizontalalignment='right', verticalalignment='center',
                  fontsize=12, fontdict={'family': 'monospace'})
marker_style = dict(color='cornflowerblue', linestyle=':', marker='o',
                    markersize=15, markerfacecoloralt='gray')


def format_axes(ax):
    ax.margins(0.2)
    ax.set_axis_off()


def nice_repr(text):
    return repr(text).lstrip('u')


fig, ax = plt.subplots()


points = []

def overlap(point, test):
    x_info = dict(p_x= point.x, t_x = test.x, Dx = point.x - test.x)
    y_info = dict(p_y= point.y, t_y = test.y, Dy = point.y - test.y)
    r_info = dict(p_r= point.r, t_r = test.r, Sum = point.r + test.r)
    dist = ((point.x - test.x)**2 +(point.y - test.y)**2)**0.5
    rs = point.r + test.r
    check = dist<rs/10
    return check

# fail=0
max_ = 100000
mrf = dict()
mr = dict()
F = []
for R in range(8):
    r = R+1
    f = 25/r
    mrf[r]= f
    F.append(f)
P = sum(F)
J = max_/P

for key, value in mrf.items():
    nv = int(np.round(value*J))
    mr[nv] = key

k = list(mr.keys())
k.sort()
k.reverse()

def max_r(i):
    for j in k:
        if i<j:
            continue
        else:
            break
    return mr[j]

for i in range(max_):
    r = randrange(max_r(i)*100, max_r(i)*100+100)/100-1

    x = 5 * random()+r/10
    y = 5 * random()+r/10
    style = marker_style
    style['markersize']= r*13
    new_point = Dot(x,y,r, style)
    for point in points:
        if overlap(point, new_point):
#             fail +=1
            break
    else:
        points.append(new_point)
        ax.plot(x,y, fillstyle='full', **style)

ax.set_aspect('equal', 'datalim')
fig.set_figwidth(7)
fig.set_figheight(7)
plt.show()
```
